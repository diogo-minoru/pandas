### Adicionando e removendo colunas

```python
import numpy as np

coffee_df["price_old"] = 4.99
coffee_df["new_price"] = np.where(coffee_df["Coffee Type"] == "Espresso", 3.99, 4.99)

coffee_df
```

```python
coffee_df.drop(columns = "price_old", inplace = True)
coffee_df
```

```python
coffee_df["revenue"] = coffee_df["Units Sold"] * coffee_df["new_price"]
coffee_df.head()
```

### Renomeando colunas

```python
coffee_df.rename(columns = {"new_price": "price"}, inplace = True)
coffee_df.head()
```

### Extraindo partes do dado

```python
bios_df["first_name"] = bios_df["name"].str.split(" ").str[0]
bios_df["last_name"] = bios_df["name"].str.split(" ").str[-1]
```

```python
bios_df["born_date"] = pd.to_datetime(bios_df["born_date"])
```

```python
bios_df["born_year"] = bios_df["born_date"].dt.year.astype("Int64")
bios_df
```

```python
bios_df["height_category"] = bios_df["height_cm"].apply(lambda x: "N/A" if pd.isnull(x) else "Short" if x < 175 else "Average" if x < 195 else "Tall")
```

```python
bios_df.sort_values("height_cm", ascending=True)
```
